```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
years = np.arange(1980, 2021) 
ages = np.arange(0, 101)     
```

```python
year_grid, age_grid = np.meshgrid(years, ages)
```

```python
year_flat = year_grid.flatten()
age_flat = age_grid.flatten()
```

```python
log_mortality_rate = -0.02 * age_flat + 0.001 * year_flat + np.random.normal(0, 0.1, size=len(year_flat))
```

```python
mortality_rate = np.exp(log_mortality_rate)
```

```python
data = pd.DataFrame({
    'Year': year_flat,
    'Age': age_flat,
    'MortalityRate': mortality_rate
})
```

```python
data.head()
```

```python
data.to_csv('simulated_mortality_rate.csv', index=False)
```
